```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

# Conjunto de puntos discretos
x = np.arange(0, 10)
y = np.exp(x)

f = interpolate.interp1d(x, y)  # función interpolación

# Visualización resultados interpolación
xc = np.arange(0, 9, 0.1)
yc = f(xc)  # Corregido: f(x) -> f(xc)

plt.plot(x, y, "o", xc, yc, "-")
plt.show()
```

```python
import numpy as np
from scipy import linalg
# Array de dimesiones 3x3
A = np.array([[8,5,5],[2,5,1],[2,7,8]])

print("\nInver de A: \n", linalg.inv(A)) # Cálculo de la matriz inversa
print("\nDterminante de A:", linalg.det(A)) # Cálculo del determinante
print("\nNorma matricial de A:", linalg.norm(A)) # Norma matricial
print("\nSeno de A:\n", linalg.sinm(A)) #Funciones trigonométricas - Seno
```
